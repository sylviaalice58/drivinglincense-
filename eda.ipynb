```python
# changing the numbers to words as 1 in gender is male and 0 is female
import pandas as pd

data['Gender'] = data['Gender'].map({
    0.0: 'Female',
    1.0: 'Male'
})

data['Gender'].value_counts()

# print(data)
```

```python
# graph of the number of males and females who are trying to apply for a license
import matplotlib.pyplot as plt
import seaborn as sns
# 1. Assign the plot to a variable named 'ax'
ax = sns.countplot(x="Gender", data=data, palette="dark")

# 2. Add the labels using the container from the plot
# Loop through all containers to label every bar automatically
for container in ax.containers:
    ax.bar_label(container)

gender_counts = data['Gender'].value_counts()
plt.title("The individuals applying for a license")

plt.xlabel('Gender')
plt.ylabel('Count')
```

```python
# changing the qualified column to words as 1 is qualified and 0 is not qualified
data['Qualified'] = data['Qualified'].map({
    0: 'Not Qualified',
    1: 'Qualified'
})
data['Qualified'].value_counts()
```

```python
# having a graph of the number of qualified and not qualified individuals/pie chart to show the percentage of qualified and not qualified individuals
import matplotlib.pyplot as plt
# Count the occurrences of each category in the 'Qualified' column
qualified_counts = data['Qualified'].value_counts()
# Create a pie chart
plt.figure(figsize=(6, 6))  
plt.pie(qualified_counts, labels=qualified_counts.index, autopct='%1.3f%%', startangle=140)
plt.title('Distribution of Qualified vs Not Qualified Individuals')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```

```python
# a linear graph precentation of qualified and not qualified individuals by age
import matplotlib.pyplot as plt

# Count the occurrences
qualified_counts = data['Qualified'].value_counts()

# Plot a line graph
plt.figure(figsize=(6, 4))

plt.plot(qualified_counts.index.astype(str), qualified_counts.values, marker='o')

plt.title('Distribution of Qualified vs Not Qualified Individuals')
plt.xlabel('Category')
plt.ylabel('Count')

plt.grid(True)

plt.show()
```

```python
# analysing the qualified and not qualified individuals by gender
pd.crosstab(data['Gender'], data['Qualified'])
```

```python
# graphical representation of the qualified and not qualified individual
import matplotlib.pyplot as plt

# Create the crosstab
ct = pd.crosstab(data['Gender'], data['Qualified'])

# Plot
ax = ct.plot(kind='bar')

plt.title('Qualified vs Not Qualified by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)

plt.legend(title='Status')

# 🔢 Add values on top of bars
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
# analysing by the percentage of qualified and not qualified individuals by gender
gender_qualified = pd.crosstab(data['Gender'], data['Qualified'], normalize='index') * 100
gender_qualified
```
